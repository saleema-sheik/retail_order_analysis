```python
1.# Find top 10 highest revenue generating products
cursor.execute('''select product_id,sum(sale_price)as sales
from order2
group by product_id
order by sales desc
limit 10;''')
table1=cursor.fetchall()
answer_1 = pd.DataFrame(table1, columns=["product_id", "sales"])

answer_1
```

```python
2.# Find the top 5 cities with the highest profit margins
cursor.execute('''
SELECT 
    order1.city,
    SUM(order2.profit) / SUM(order2.sale_price) AS profit_margin
FROM 
    order1 
JOIN 
    order2 ON order1.order_id = order2.order_id
GROUP BY 
    order1.city
ORDER BY 
    profit_margin DESC
LIMIT 5;''')
table2=cursor.fetchall()
answer_2 = pd.DataFrame(table2, columns=["city", "profit_margin"])

answer_2 
```

```python
3.# Calculate the total discount given for each category
cursor.execute('''
SELECT 
    order1.category,
    SUM(order2.discount) AS total_discount
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order1.category
ORDER BY 
    total_discount DESC;''')
table3=cursor.fetchall()
answer_3 = pd.DataFrame(table3, columns=["category", "total_discount"])

answer_3
```

```python
4.# Find the average sale price per product category
cursor.execute('''
SELECT 
    sub_category,
    AVG(sale_price) AS average_sale_price
FROM 
    order2 
GROUP BY 
    sub_category
ORDER BY 
    average_sale_price DESC;''')
table4=cursor.fetchall()
answer_4 = pd.DataFrame(table4, columns=["sub_category", "average_sale_price"])

answer_4

```

```python
5.# Find the region with the highest average sale price
cursor.execute('''
SELECT 
    order1.region,
    AVG(order2.sale_price) AS average_sale_price
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order1.region
ORDER BY 
    average_sale_price DESC;''')
table5=cursor.fetchall()
answer_5 = pd.DataFrame(table5, columns=["region", "average_sale_price"])

answer_5
```

```python
6.# Find the total profit per category
cursor.execute('''
SELECT 
    order1.category,
    SUM(order2.profit*quantity) AS total_profit
FROM 
    order1 
JOIN 
    order2 ON order1.order_id = order2.order_id
GROUP BY 
    category
ORDER BY 
    total_profit DESC;''')
table6=cursor.fetchall()
answer_6 = pd.DataFrame(table6, columns=["category", "total_profit"])

answer_6
```

```python
7.# Identify the top 3 segments with the highest quantity of orders
cursor.execute('''
SELECT 
    order1.segment,
    SUM(order2.quantity) AS highest_quantity
FROM 
    order1 
JOIN 
    order2 ON order1.order_id = order2.order_id
GROUP BY 
    segment
ORDER BY 
    highest_quantity DESC;''')
table7=cursor.fetchall()
answer_7 = pd.DataFrame(table7, columns=["segment", "highest_quantity"])

answer_7
```

```python
8.# Determine the average discount percentage given per region

cursor.execute('''
SELECT 
    order1.region,
    AVG(order2.discount_percent) AS average_discount_percentage
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order1.region
ORDER BY 
    average_discount_percentage DESC;''')
table8=cursor.fetchall()
answer_8 = pd.DataFrame(table8, columns=["region", "average_discount_percentage"])

answer_8

```

```python
9.# Find the product category with the highest total profit
cursor.execute('''
SELECT 
    sub_category,
    SUM(profit*quantity) AS total_profit
FROM 
    order2 
GROUP BY 
    sub_category 
ORDER BY 
    total_profit DESC;''')
table9=cursor.fetchall()
answer_9 = pd.DataFrame(table9, columns=["sub_category", "total_profit"])

answer_9
```

```python
10.# Calculate the total revenue generated per year
cursor.execute('''
SELECT 
    YEAR(order1.order_date) AS order_year,
    SUM(order2.sale_price *quantity) AS total_revenue
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order_year
ORDER BY 
    order_year;''')
table10=cursor.fetchall()
answer_10 = pd.DataFrame(table10, columns=["order_year", "total_revenue"])

answer_10

```

```python
11.#Calculate the Total Number of Orders per City
cursor.execute('''
SELECT 
    order1.city,
    COUNT(DISTINCT order1.order_id) AS total_orders
FROM 
    order1 
GROUP BY 
    order1.city
ORDER BY 
    total_orders DESC;''')
table11=cursor.fetchall()
answer_11 = pd.DataFrame(table11, columns=["city", "total_orders"])

answer_11

```

```python
12.#calculate the total discount per state
cursor.execute('''
SELECT 
    order1.state,
    SUM(order2.discount) AS total_discount
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order1.state
ORDER BY 
    total_discount DESC;

''')
table12=cursor.fetchall()
answer_12 = pd.DataFrame(table12, columns=["state", "tottal_discount"])

answer_12

```

```python
13.#List the Top 3 Cities with the Highest Total Sales
cursor.execute('''
SELECT 
    order1.city,
    SUM(order2.sale_price * order2.quantity) AS total_sales
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order1.city
ORDER BY 
    total_sales DESC
LIMIT 3;''')
table13=cursor.fetchall()
answer_13 = pd.DataFrame(table13, columns=["city", "total_sales"])

answer_13
```

```python
14.#Determine the Number of Unique Products Sold per Region
cursor.execute('''
SELECT 
    order1.region,
    COUNT(DISTINCT order2.product_id) AS unique_products_sold
FROM 
    order1 
JOIN 
    order2  ON order1.order_id = order2.order_id
GROUP BY 
    order1.region;
''')
table14=cursor.fetchall()
answer_14 = pd.DataFrame(table14, columns=["Region", "unique_products_sold"])

answer_14
```

```python
15.#Determine total profit for each year
cursor.execute('''

SELECT 
     YEAR(order1.order_date) AS order_year,
     sum(order2.profit) AS total_profit
FROM 
    order1
JOIN
    order2 ON order1.order_id = order2.order_id
GROUP BY
    order_year;
''')
table15=cursor.fetchall()
answer_15 = pd.DataFrame(table15, columns=["order_year", "total_profit"])

answer_15
```

```python
16.#Average Order Value per Shipping Mode
cursor.execute('''
SELECT 
     order1.ship_mode, AVG(order2.sale_price * order2.quantity) AS average_order_value
FROM 
    order1 
JOIN 
    order2 ON order1.order_id = order2.order_id
GROUP BY 
     order1.ship_mode
ORDER BY 
      average_order_value DESC;
''')

table16=cursor.fetchall()
answer_16 = pd.DataFrame(table16, columns=["ship_mode", "average_order_value"])

answer_16
```

```python
17.#Total Quantity Sold per State

cursor.execute('''
SELECT 
     order1.state, SUM(order2.quantity) AS total_quantity_sold
FROM 
     order1
JOIN 
     order2 ON order1.order_id = order2.order_id
GROUP BY 
     order1.state
ORDER BY 
     total_quantity_sold DESC;

''')
table17=cursor.fetchall()
answer_17 = pd.DataFrame(table17, columns=["state", "total_quantity_sold"])

answer_17
```

```python
18.#Top 5 Sub-Categories by Profit

cursor.execute('''
SELECT 
     order2.sub_category, SUM(order2.profit) AS total_profit
FROM 
     order2
GROUP BY 
     order2.sub_category
ORDER BY 
      total_profit DESC
LIMIT 5;

''')
table18=cursor.fetchall()
answer_18 = pd.DataFrame(table18, columns=["sub_category", "total_profit"])

answer_18
```

```python
19.#Monthly Sales Trend for a Specific Year

cursor.execute('''
SELECT 
     MONTH(order1.order_date) AS month, SUM(order2.sale_price * order2.quantity) AS monthly_sales
FROM 
     order1 
JOIN 
     order2 ON order1.order_id = order2.order_id
WHERE 
     YEAR(order1.order_date) = 2023
GROUP BY 
     MONTH(order1.order_date)
ORDER BY 
      month;

''')
table19=cursor.fetchall()
answer_19 = pd.DataFrame(table19, columns=["month", "monthly_sales_2023"])

answer_19
```
